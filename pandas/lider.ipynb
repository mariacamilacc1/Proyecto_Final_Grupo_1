```python
df = pd.read_csv('Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv', encoding='utf8') 
```

```python
df
```

```python
def sumadoc(n):
    n1=str(n["Número de documento"])
    s=0
    for i in n1:
        s+=int(i)
    return s
```

```python
df["N"]=df.apply(sumadoc, axis=1)
```

```python
def fcosto1(n):
    n1=int(n[4]) 
    c105=(n1*105**2)+(2*105)+23
    c100=(n1*100**2)+(2*100)+23
    fc1=(c105-c100)/(105-100)
    return fc1
```

```python
df["Punto 1"]=df.apply(fcosto1, axis=1)
```

```python
def calificación1(n):
    n1=int(n[5])
    n2=int(n[1])
    if n1==n2:
        return 5
    else:
        return 1
```

```python
df["Calificación P1"]=df.apply(calificación1, axis=1)
```
