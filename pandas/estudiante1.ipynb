```python
def c2(x):
    n=int(x[4])
    s= n*101**2+2*101+23
    s1=n*100**2+2*100+23
    sr=(s-s1)
    return float(sr)
```

```python
dp["Punto 2"]=dp.apply(c2,axis=1)
```

```python
def cal2(x):
    n=int(x[2])
    n1=int(x[7])
    if n!=n1:
        return 1
    else:
        return 5
```

```python
dp["calificacion 2"]=dp.apply(cal2,axis=1)
```
